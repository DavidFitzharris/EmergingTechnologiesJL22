<p>A signal can be broken down into its sine and cosine components using a mathematical technique called the Fourier transform.
A multitude of applications, including digital signal processing, communications systems, radar systems, etc., require the ability to analyze signals in both the time and frequency domain.
The Fourier transform uses an algorithm that separates a signal's frequencies and amplitudes into constituent parts.
We can better comprehend the properties of the signal being analyzed by looking at these component frequencies and amplitudes. </p>

<h2 style="color: rgb(0, 120, 60)">Sinusoidal</h2>
<hr style="border-top: 1px dashed rgb(0, 120, 60)" />

<p>The term given to describe a sine wave, a curve that moves in periodic oscillation, is sinusoidal.<br>
    Below we can look at creating a simple sinusoid in python

```python
import numpy as np

import matplotlib.pyplot as plt


## sample rate
Fs = 6000

## Signal Frequency
f = 4

## x axis
sample = 6000
x = np.arange(sample)

## Calculate value of sin wave
y = np.sin(2 * np.pi * f * x / Fs)

## plot graph
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
